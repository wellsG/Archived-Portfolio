### House Sales Analysis using python

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
file_name='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/coursera/project/kc_house_data_NaN.csv'
df=pd.read_csv(file_name)
```

### Data Wrangling

```python
#Searching for missing data values
df.drop(columns=['id','Unnamed: 0'], axis=1, inplace=True)
df.describe()
```

```python
#Determine the number of null values
print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```

```python
#Replace null with mean for bedrooms
mean=df['bedrooms'].mean()
df['bedrooms'].replace(np.nan,mean, inplace=True)
```

```python
#Replace null with mean for bathrooms
mean=df['bathrooms'].mean()
df['bathrooms'].replace(np.nan,mean, inplace=True)
```

### Data Exploration

### Model Development

```python
#Fitting 'long' to 'price' using linear regression method
X = df[['long']]
Y = df['price']
lm = LinearRegression()
lm.fit(X,Y)
lm.score(X, Y)
```

```python
#Linear regression to predit price in relation to square foot living
Y = df[['price']]
X = df[['sqft_living']]
lm = LinearRegression()
lm.fit(X,Y)
lm.score(X, Y)
```

```python
#lr fitting features to price
features = df[["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]]
Y = df['price']
lm = LinearRegression()
lm.fit(features, Y)
```

```python
#R2 value of features to price
lm.score(features, Y)
```

```python
#Establishign pipeline Input for model refinement
Input=[('scale',StandardScaler()),('polynomial', PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
```

```python
#Fitting data to pipeline
pipe=Pipeline(Input)
pipe
pipe.fit(features, Y)
pipe.score(features, Y)
```

### Model Refinement

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
print("done")
```

```python
#Defining variables to train & test samples
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)


print("number of test samples:", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
from sklearn.linear_model import Ridge
```

```python
RR=Ridge(alpha=0.1)
RR.fit(x_train,y_train)
RR.score(x_train, y_train)
```

```python
#2nd order polynomial transformation
pr=PolynomialFeatures(degree=2)
x_train_pr=pr.fit_transform(x_train)
x_test_pr=pr.fit_transform(x_test)
RR=Ridge(alpha=0.1)
RR.fit(x_train_pr,y_train)
RR.score(x_test_pr, y_test)
```
